```
import pandas as pd
d=pd.read_csv('sleepdataprocessed.csv') 
t=[]
short_data = d.copy()
t.extend(short_data["Sleep_quality_per"].tolist())


for i in range(0,len(t)):
  t[i]=int(t[i]*100)
b=[]
bad=0
avg=0
best=0
for i in t:
  if(i>=0 and i<=40):
    bad=bad+1
  elif(i>40 and i<=70):
    avg=avg+1
  else:
    best=best+1



import matplotlib.pyplot as plt
fig=plt.figure(figsize=(13,10), dpi= 50)
ax = fig.add_axes([0,0,1,1])
plt.xlabel("sleep quality")
plt.ylabel("sleep data")
plt.title("Figure to illustrate the Sleep stages")
langs = ['bad', 'avg', 'best']
students = [bad,avg,best]
ax.bar(langs,students, color='black')

total=sum(students)
k=0
for i in ax.patches:
    # get_x pulls left or right; get_height pushes up or down
    ax.text(i.get_x()-.01, i.get_height()+.6,'        '+str((round(students[k]*100/total,1)))+'%', fontsize=28,color='dimgrey')
    k=k+1


plt.show()

```

```
import pandas as pd
from sklearn.ensemble  import RandomForestClassifier
# Read in data and display first 5 rows
df = pd.read_csv('sleepdataprocessed.csv')
df.head(5)
df=df.drop(['Sleep_quality','Activity_steps'],axis=1)




X = df.drop('Stressful_Day', axis=1)
y = df['Stressful_Day']

from sklearn.model_selection import train_test_split
# implementing train-test-split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=66)


from sklearn import model_selection
# random forest model creation
rfc = RandomForestClassifier()
rfc.fit(X_train,y_train)
# predictions
rfc_predict = rfc.predict(X_test)

from sklearn.model_selection import cross_val_score
from sklearn.metrics import classification_report, confusion_matrix


rfc_cv_score = cross_val_score(rfc, X, y, cv=10, scoring='roc_auc')



print("=== Confusion Matrix ===")
print(confusion_matrix(y_test, rfc_predict))

print('\n')
print("=== Classification Report ===")
print(classification_report(y_test, rfc_predict))
print('\n')
print("=== All AUC Scores ===")
print(rfc_cv_score)
print('\n')
print("=== Mean AUC Score ===")
print("Mean AUC Score - Random Forest: ", rfc_cv_score.mean())


```

```
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test, rfc_predict)


import seaborn as sn
import pandas as pd
import matplotlib.pyplot as plt

array=confusion_matrix(y_test,rfc_predict)
print(array)
df_cm = pd.DataFrame(array, range(2), range(2))
plt.figure(figsize=(7,5))
sn.set(font_scale=1.4) # for label size
sn.heatmap(df_cm, annot=True, annot_kws={"size": 18}) # font size

plt.show()
```

```
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

lin=pd.read_csv(r"sleepdataprocessed.csv")
lin.head()
```

```
X=lin.drop("Sleep_quality",axis=1)
y=lin['Sleep_quality']


from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test=train_test_split(X, y, test_size=0.25, random_state=0)


from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
X_train=sc.fit_transform(X_train)
X_test=sc.transform(X_test)


import time
from sklearn.naive_bayes import GaussianNB
NB=GaussianNB()
NB.fit(X_train, y_train)
training_start = time.perf_counter()
training_end = time.perf_counter()
prediction_start = time.perf_counter()
prediction_end = time.perf_counter()
rfc_train_time = training_end-training_start
rfc_prediction_time = prediction_end-prediction_start
print("Time consumed for training: %4.3f seconds" % (rfc_train_time))
print("Time consumed for prediction: %6.5f seconds" % (rfc_prediction_time))

```

```
preds=NB.predict(X_test)
preds
```

```
from sklearn.metrics import classification_report
classification_report(y_test, preds)
```

```
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test, preds)



import seaborn as sn
import pandas as pd
import matplotlib.pyplot as plt

array=confusion_matrix(y_test,preds)
print(array)
df_cm = pd.DataFrame(array, range(2), range(2))
plt.figure(figsize=(7,5))
sn.set(font_scale=1.4) # for label size
sn.heatmap(df_cm, annot=True, annot_kws={"size": 18}) # font size

plt.show()

```

```
import pandas as pd
d=pd.read_csv('sleepdataprocessed.csv') 
t=[]
q=['Sleep_quality_per','Total_time_in_bed_mins','drank_Cofee','Drank_Tea','Worked_Out','Stressful_Day','Ate_late']
short_data = d.head(800).copy()
t.append(short_data["Sleep_quality_per"].tolist())
t.append(short_data["Total_time_in_bed_mins"].tolist())
t.append(short_data["drank_Cofee"].tolist())
t.append(short_data["Drank_Tea"].tolist())
t.append(short_data["Worked_Out"].tolist())
t.append(short_data["Stressful_Day"].tolist())
t.append(short_data["Ate_late"].tolist())

for i in range(0,len(t[0])):
  t[0][i]=int(t[0][i]*100)




import matplotlib.pyplot as plt
plt.rc('figure', figsize=[12,7])

sy=[int(t[0][i]) for i in range(0,len(t[0])) if t[2][i]==1]
ty=[int(t[1][i]) for i in range(0,len(t[1])) if t[2][i]==1]
sn=[int(t[0][i]) for i in range(0,len(t[0])) if t[2][i]==0]
tn=[int(t[1][i]) for i in range(0,len(t[1])) if t[2][i]==0]

plt.scatter(sy, ty, marker='o', color=['red'],label='Drank Cofee')
plt.scatter(sn, tn, marker='x', color=['red'],label='Drank Cofee')
plt.xlabel('Sleep quality', fontsize=16)
plt.ylabel('Total time in bed', fontsize=16)
plt.title('Sleep data analysis for '+str(q[2]),fontsize=20)
plt.legend()
plt.show()



```
